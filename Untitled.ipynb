```python
# column chest , totlengh , hdlengh is good correlation so i will drop given column
```

```python
df.columns
```

```python
df.drop(["chest","hdlngth","totlngth"],axis = 1 ,inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df["Pop"] = np.where(df["Pop"]=="Vic",1,0)
```

```python
df.head()
```

```python
x = df.drop("sex",axis = 1)
y =df["sex"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=65,stratify=y)
```

```python
# model training
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
log_model = LogisticRegression()
log_model.fit(x_train,y_train)
```

```python
# training evaluation matrix
y_pred_train = log_model.predict(x_train)
acc = accuracy_score(y_train,y_pred_train)
print("accuracy_score>>",acc)
conf = confusion_matrix(y_train,y_pred_train)
print("confusion_matrix\n>>",conf)
clas = classification_report(y_train,y_pred_train)
print("classification_report\n>>",clas)
```

```python
# testing evaluation matrix
y_pred_test = log_model.predict(x_test)
acc = accuracy_score(y_test,y_pred_test)
print("accuracy_score>>",acc)
conf = confusion_matrix(y_test,y_pred_test)
print("confusion_matrix\n>>",conf)
clas = classification_report(y_test,y_pred_test)
print("classification_report\n>>",clas)
```

```python
# over logistic regression got good accuracy but this accuracy isnt accept so i will go next.
```

```python
from sklearn.linear_model import RidgeClassifier
```

```python
rc_model = RidgeClassifier(alpha=3.1)
rc_model.fit(x_train,y_train)
```

```python
# training evaluation matrix
y_pred_train = rc_model.predict(x_train)
acc = accuracy_score(y_train,y_pred_train)
print("accuracy_score>>",acc)
conf = confusion_matrix(y_train,y_pred_train)
print("confusion_matrix\n>>",conf)
clas = classification_report(y_train,y_pred_train)
print("classification_report\n>>",clas)
```

```python
# testing evaluation matrix
y_pred_test = rc_model.predict(x_test)
acc = accuracy_score(y_test,y_pred_test)
print("accuracy_score>>",acc)
conf = confusion_matrix(y_test,y_pred_test)
print("confusion_matrix\n>>",conf)
clas = classification_report(y_test,y_pred_test)
print("classification_report\n>>",clas)
```

```python
# over ridge got little good accuracy but i need to tuned the model
```

```python
rc_model = RidgeClassifier()
hyp = {"alpha":np.arange(0.1,10)}
rscv_model = RandomizedSearchCV(rc_model,hyp,cv = 5)
rscv_model.fit(x_train,y_train)
rscv_model.best_estimator_
```

```python
# Same over ridge classifier not got good accuracy we will moved to nexr
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier(min_samples_leaf=2, min_samples_split=11)
dt_model.fit(x_train,y_train)
```

```python
# training evaluation matrix
y_pred_train = dt_model.predict(x_train)
acc = accuracy_score(y_train,y_pred_train)
print("accuracy_score>>",acc)
conf = confusion_matrix(y_train,y_pred_train)
print("confusion_matrix\n>>",conf)
clas = classification_report(y_train,y_pred_train)
print("classification_report\n>>",clas)
```

```python
# testing evaluation matrix
y_pred_test = dt_model.predict(x_test)
acc = accuracy_score(y_test,y_pred_test)
print("accuracy_score>>",acc)
conf = confusion_matrix(y_test,y_pred_test)
print("confusion_matrix\n>>",conf)
clas = classification_report(y_test,y_pred_test)
print("classification_report\n>>",clas)
```

```python
dt_model = DecisionTreeClassifier()
hyp = {"criterion":["gini","entropy"],"min_samples_split":np.arange(2,20),
      "min_samples_leaf":np.arange(2,15),'min_samples_leaf':np.arange(1,10)}
rscv_model = RandomizedSearchCV(dt_model,hyp,cv = 5)
rscv_model.fit(x_train,y_train)
rscv_model.best_estimator_
```

```python
# Over decision tree got good accuracy
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(max_features='auto', min_samples_leaf=3,
                       min_samples_split=8, n_estimators=41, oob_score=True)
rf_model.fit(x_train,y_train)
```

```python
# training evaluation matrix
y_pred_train = rf_model.predict(x_train)
acc = accuracy_score(y_train,y_pred_train)
print("accuracy_score>>",acc)
conf = confusion_matrix(y_train,y_pred_train)
print("confusion_matrix\n>>",conf)
clas = classification_report(y_train,y_pred_train)
print("classification_report\n>>",clas)
```

```python
# testing evaluation matrix
y_pred_test = rf_model.predict(x_test)
acc = accuracy_score(y_test,y_pred_test)
print("accuracy_score>>",acc)
conf = confusion_matrix(y_test,y_pred_test)
print("confusion_matrix\n>>",conf)
clas = classification_report(y_test,y_pred_test)
print("classification_report\n>>",clas)
```

```python
rf_model  = RandomForestClassifier()
hyp = {"n_estimators":np.arange(1,100),"max_features":["auto","log2"],
       "bootstrap":[True],"oob_score":[True],
    "criterion":["gini","entropy"],"min_samples_split":np.arange(2,20),
      "min_samples_leaf":np.arange(2,15),'min_samples_leaf':np.arange(1,10)}
rscv_model = RandomizedSearchCV(rf_model,hyp,cv = 5)
rscv_model.fit(x_train,y_train)
rscv_model.best_estimator_
```
